## (b) What is the *range* of each quantitative predictor? You can answer this using the `range()` function.

```python
as.data.frame(sapply(Auto[-c(9)], range))
```

## (c) What is the mean and standard deviation of each quantitative predictor?

```python
mean <- sapply(Auto[-c(9)], mean)
std <- sapply(Auto[-c(9)], sd)
df <- as.data.frame(mean)
df["std"] <- std
df
```

As we can see from the scatterplot and the correlation matrix,
* if the plot looks like the graph of $y = x$, they are **positively correlated**.
* if the plot seems like the graph of $y = 1/x$, they are **negatively correlated**.
* otherwise, if it shows no particular direction, their correlation is **weak**.
